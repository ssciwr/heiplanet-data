```python
from tinydb import TinyDB, Query
```

```python
db = TinyDB("db.json")
Query = Query()
```

```python
data_example = {
    "signature": "hash-signature-data-abc123",
    "signature_data": {
        "ds_name": "dataset_1",
        "product_type": "reanalysis",
        "data_var": "t2m",
    },
    "year": ["2022", "2023", "2024"],
    "month": ["1", "2", "3"],
    "time": ["00:00"],
    "file_path": "/data/dataset_1/t2m/2022/1/dataset_1_t2m_1.nc",
    "data_format": "netcdf",
    "download_format": "unarchived",
    "downloaded_at": "2024-06-15T12:00:00",
    "status": "active",
}
```

```python
example_1 = data_example.copy()
example_1["signature"] = data_example["signature"] + "-t2m"
example_1["file_path"] = "/data/dataset_1/t2m/2022/1/dataset_1_t2m_1_v2.nc"
example_2 = data_example.copy()
example_2["signature"] = data_example["signature"] + "-tp"
example_2["signature_data"]["data_var"] = "tp"
example_2["file_path"] = "/data/dataset_1/tp/2022/1/dataset_1_tp_1_v3.nc"
```

```python
db.insert_multiple([example_1, example_2])
```

```python
db.all()
```

```python
# get all file paths of dataset_1
results = db.search(Query.signature_data.ds_name == "dataset_1")
for r in results:
    print(r["file_path"])
```

```python
for item in results:
    print(type(item))
```

```python
# find file paths that have data in 2024
results = db.search(Query.year.any("2024"))
for r in results:
    print(r["file_path"])
```

```python
example_3 = data_example.copy()
example_3["signature"] = data_example["signature"] + "-t2m"
example_3["file_path"] = "/data/dataset_1/t2m/2023/2/dataset_1_t2m_2_v1.nc"
example_3["year"] = ["2024"]
example_3["month"] = ["4", "5"]
db.insert(example_3)
```

```python
# find all files that have t2m in 2024 from Jan to Apr
results = db.search(
    (Query.signature == "hash-signature-data-abc123-t2m")
    & Query.year.any(["2024"])
    & Query.month.any(["1", "2", "3", "4"])
)
for r in results:
    print(r["file_path"])
```
